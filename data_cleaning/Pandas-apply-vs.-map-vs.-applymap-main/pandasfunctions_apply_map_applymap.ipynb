```python
# df.apply
# # if you apply the .apply function to a whole dataframe you pass one column at a time or pass one row at a time
# # if you apply the .apply function on a column of a dataframe you are passing one element at a time only in one column
# # you can apply the .map function to a series object or certain column in dataframe the function that you are passing will be applied to each element of the specified column
# # you can only use the df.applymap function on whole dataframes
```

```python
import pandas as pd
```

```python
data = pd.read_csv("NYC_Jobs.csv")
data.head()
```

```python
def capitalize(title):
    return title.lower().capitalize()

data['Civil Service Title'].apply(capitalize)
```

```python
def add_year(text):
    return str(text) + '_2022'

data.applymap(add_year)
```

```python
# A Pandas Series is a one-dimensional labeled array in the Pandas library for Python. It is capable of holding any data type, including integers, strings, floating-point numbers, and Python objects. 

s = pd.Series(['cat', 'dog', 'NaN', 'rabbit'])
s
```

```python
# # .map is able to find occurences of a certain str or element and changing it immediately
# # strength of .map is if you want to change something quickly you can pass a dictionary to a map function and it will be done
# # if you try to do the same with the .apply function you would have to create a funtion to check each element to see if it changed the str or element

s.map({'cat':'kitten', 'dog':'puppy'})
```

```python
# # you can also use .map using functions
# # you can either pass a dictionary to a .map function or a whole function, but you can do the same thing with .apply
# # if there is no match then .map will turn things into a missing value

def change_word(title):
    return title.replace('DEPT', 'kitten')

data['Agency'].map(change_word)
```

```python
# # use .map to add something to a whole column
# # if you want to apply something to just one column and not create a new function to do the function operation and just want to change the values

data['Agency'].map('This position is offered by {} '.format)
```

```python
# # applies to all of these functions
# # when you are passing a function to them and the function cannot handle (example) 'NaN' or missing values/nonvalues

# # if there is not a number value your function may break if 'NaN' # using .ignore is essentially saying to not even pass the 'NaN' values
# s.map(function, na_action='ignore')

s.map({'cat': 'kitten', 'dog': 'puppy'}, na_action='ignore')
```
