```python
hotel_bookings['agent'].unique() # In the context of the pandas library in Python, "unique" refers to the distinct values within a Series (a single column of a DataFrame) or an Index.
hotel_bookings[hotel_bookings['agent'] == 5]

hotel_bookings['agent'].fillna(-1, inplace=True)

hotel_bookings[hotel_bookings['num_children'].isna()] # "NA" means "not available" - the .isna() method is used to detect missing values in a dataframe or series 

hotel_bookings[hotel_bookings['country'].isna()] [['hotel', 'is_canceled', 'country']]

hotel_bookings['country'].fillna('Unknown', inplace=True)

hotel_bookings.dropna(subset=['num_children'], inplace=True)
```

```python
hotel_bookings = hotel_bookings.astype({'is_canceled': 'boolean', 'is_repeated_guest': 'boolean', 'num_children': 'int', 'agent': 'int'}) # Changed first two to boolean and last two to integers 
# The astype() function in Pandas is used to cast a Pandas object (like a DataFrame or Series) to a specified data type (dtype). This method is fundamental for data cleaning and preparation, ensuring that columns have the correct data types for subsequent analysis or operations. 
hotel_bookings.dtypes
```

```python
# Bin columns # Creating "bins" in pandas refers to the process of discretizing or categorizing continuous numerical data into discrete intervals or groups. This technique is also known as binning or bucketing.
hotel_bookings['lead_time'].unique() # In a data context using the pandas library, "lead time" refers to the time elapsed between two defined events or timestamps recorded in your dataset. It is not a built-in function but a concept you calculate by taking the difference between two datetime columns. 
# This is a common operation in time-series analysis for inventory management, project tracking, and other time-based workflows.
hotel_bookings['lead_time'].describe() # The .describe() function in the pandas library for Python is used to generate descriptive statistics of a DataFrame or a Series. It provides a quick summary of the central tendency, dispersion, and shape of a dataset's distribution, excluding NaN (Not a Number) values.

bins = [0, 100, 200, 300, 400, 500, 600, 700, 800] # The pd.cut function in the pandas library is used to bin values into discrete intervals. It is primarily used to convert a continuous numerical variable into a categorical one by segmenting and sorting data values into specified bins.
labels = ['0-100', '101-200', '201-300', '301-400', '401-500', '501-600', '601-700', '701-800'] # here the pd.cut looks at the lead time value and will assign it one of the labels based on which bin it falls under 
# In Pandas, "labels" refer to the named identifiers used to access and organize data within Series and DataFrames. These labels are crucial for label-based indexing, data alignment, and providing metadata about the data. 
# The values on above line are the bins (eg. '101-200')

hotel_bookings['lead_time_binned'] = pd.cut(hotel_bookings['lead_time'], bins=bins, labels=labels) # breakdown of actions = pd.cut(specify the column to cut) |hotel_bookings['lead_time']|, (specify bins) |bins=bins| (specify labels) |labels=labels|

hotel_bookings[['lead_time', 'lead_time_binned']]
```

```python
# String cleaning

hotel_bookings['hotel'] = hotel_bookings['hotel'].replace(r"[\*\n\^]", '', regex=True)    # .str.contains(pattern, regex=True) → checks if text matches pattern    # .str.replace(pattern, repl, regex=True) → replaces matches with something else    # .str.extract(pattern) → extracts part of the string that matches
hotel_bookings['hotel'].unique()
```

```python
# Remove duplicates

# In pandas, .duplicated() is a method available for both Series and DataFrames that identifies duplicate values or rows. It returns a boolean Series or array of the same length as the original object, where True indicates a duplicate and False indicates a unique value/row.
# 'keep' - the concept of "keep" generally refers to retaining specific data elements within a DataFrame or Series, often in the context of filtering or removing duplicates.
# 'keep' - whether or not you want to count the first or last occurence of a duplicated row as the duplicated row
# if you set 'keep' to first it will return that the first instance of a row is not a duplicate, but every other time you see it, it will be marked as True
# if you set 'keep' to False where each instance of the duplicated row is treated as a duplicated row and this function returns True for that row
# 'keep' is a function in '.duplicated()'

hotel_bookings.loc[hotel_bookings.duplicated(keep=False)]

# don't need to keep both rows, just the first instance of each row and drop the rest of them, so below line is used
hotel_bookings.drop_duplicates(keep="first", inplace=True) 
# takes a subset parameter where you can look for duplicates just within certain columns
# if you don't specify this it will drop duplicates where the values for all of the columns are the same and that is what we want to do here
```
